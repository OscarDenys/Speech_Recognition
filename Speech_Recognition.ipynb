```python
from torch.serialization import INT_SIZE
class Network(torch.nn.Module):
    def __init__(self, context):
        super(Network, self).__init__()
        # TODO: Please try different architectures
        in_size = 13*(1+2*context)
        layers = [
            nn.Linear(in_size, 1024),

            nn.BatchNorm1d(1024),
            nn.ReLU(),
            nn.Dropout(p=0.1),

            nn.Linear(1024, 2048),

            nn.BatchNorm1d(2048),
            nn.ReLU(),
            nn.Dropout(p=0.1),

            nn.Linear(2048,4096),

            nn.BatchNorm1d(4096),
            nn.ReLU(),
            nn.Dropout(p=0.1),

            nn.Linear(4096, 2048),

            nn.BatchNorm1d(2048),
            nn.ReLU(),
            nn.Dropout(p=0.1),

            nn.Linear(2048, 1024),

            nn.BatchNorm1d(1024),
            nn.ReLU(),
            nn.Dropout(p=0.1),

            nn.Linear(1024,512),
            
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(p=0.1),

            nn.Linear(512,256),

            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(p=0.1),

            nn.Linear(256,128),

            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(p=0.1),
            
            nn.Linear(128,40)
        ]
        self.laysers = nn.Sequential(*layers)

    def forward(self, A0):
        x = self.laysers(A0)
        return x
```

```python

class LibriSamples(torch.utils.data.Dataset):
    def __init__(self, data_path, sample=20000, shuffle=True, partition="dev-clean", csvpath=None, predict=False):
        # sample represent how many npy files will be preloaded for one __getitem__ call
        self.sample = sample 
        # predict is a Boolean indicating that the samples will be used on test-clean (no ground truth available)
        self.predict = predict
        
        self.X_dir = data_path + "/" + partition + "/mfcc/"
        if not predict:
          self.Y_dir = data_path + "/" + partition +"/transcript/"
        
        self.X_names = os.listdir(self.X_dir)
        if not predict:
          self.Y_names = os.listdir(self.Y_dir)

        # using a small part of the dataset to debug
        if csvpath:
            if not predict:
                subset = self.parse_csv(csvpath)
                self.X_names = [i for i in self.X_names if i in subset]
                self.Y_names = [i for i in self.Y_names if i in subset]
            else:
                self.X_names = list(pd.read_csv(csvpath).file)
        
        if (shuffle == True) and (not predict):
            XY_names = list(zip(self.X_names, self.Y_names))
            random.shuffle(XY_names)
            self.X_names, self.Y_names = zip(*XY_names)
        
        if not predict:
          assert(len(self.X_names) == len(self.Y_names))
        self.length = len(self.X_names)
        
        self.PHONEMES = [
            'SIL',   'AA',    'AE',    'AH',    'AO',    'AW',    'AY',  
            'B',     'CH',    'D',     'DH',    'EH',    'ER',    'EY',
            'F',     'G',     'HH',    'IH',    'IY',    'JH',    'K',
            'L',     'M',     'N',     'NG',    'OW',    'OY',    'P',
            'R',     'S',     'SH',    'T',     'TH',    'UH',    'UW',
            'V',     'W',     'Y',     'Z',     'ZH',    '<sos>', '<eos>']
      
    @staticmethod
    def parse_csv(filepath):
        subset = []
        with open(filepath) as f:
            f_csv = csv.reader(f)
            for row in f_csv:
                subset.append(row[1])
        return subset[1:]

    def __len__(self):
        return int(np.ceil(self.length / self.sample))
        
    def __getitem__(self, i):
        sample_range = range(i*self.sample, min((i+1)*self.sample, self.length))
        if not self.predict:
          X, Y = [], []
          for j in sample_range:
              X_path = self.X_dir + self.X_names[j]
              Y_path = self.Y_dir + self.Y_names[j]
              
              label = [self.PHONEMES.index(yy) for yy in np.load(Y_path)][1:-1]

              X_data = np.load(X_path)
              X_data = (X_data - X_data.mean(axis=0))/X_data.std(axis=0)
              X.append(X_data)
              Y.append(np.array(label))
              
          X, Y = np.concatenate(X), np.concatenate(Y)
          return X, Y
        if self.predict:
          X = []
          for j in sample_range:
              X_path = self.X_dir + self.X_names[j]
              X_data = np.load(X_path)
              X_data = (X_data - X_data.mean(axis=0))/X_data.std(axis=0)
              X.append(X_data)
              
          X = np.concatenate(X)
          return X
    
class LibriItems(torch.utils.data.Dataset):
    def __init__(self, X, Y=0, context = 0, predict = False):
        if not predict:
          assert(X.shape[0] == Y.shape[0])
        
        self.length  = X.shape[0]
        self.context = context
        self.predict = predict

        if context == 0:
            self.X, self.Y = X, Y
        else:
            X = np.pad(X, ((context,context), (0,0)), 'constant', constant_values=(0,0))
            self.X, self.Y = X, Y
        
    def __len__(self):
        return self.length
        
    def __getitem__(self, i):
        if (self.context == 0) and (not self.predict):
            xx = self.X[i].flatten()
            yy = self.Y[i]
        if (self.context == 0) and (self.predict):
            xx = self.X[i].flatten()
        elif not self.predict:
            xx = self.X[i:(i + 2*self.context + 1)].flatten()
            yy = self.Y[i]
        elif self.predict:
            xx = self.X[i:(i + 2*self.context + 1)].flatten()

        if not self.predict:
          return xx, yy
        else:
          return xx
```

```python

def train(args, model, device, train_samples, optimizer, criterion, epoch):
    model.train()
    scaler = torch.cuda.amp.GradScaler()

    for i in range(len(train_samples)):
        X, Y = train_samples[i]
        train_items = LibriItems(X, Y, context=args['context'])
        train_loader = torch.utils.data.DataLoader(train_items, batch_size=args['batch_size'], shuffle=True)

        for batch_idx, (data, target) in enumerate(train_loader):
            data = data.float().to(device)
            target = target.long().to(device)
            optimizer.zero_grad()

            with torch.cuda.amp.autocast():
              output = model(data)
              loss = criterion(output, target)

            # loss.backward()
            scaler.scale(loss).backward()
            scaler.step(optimizer)

            # optimizer.step()
            scaler.update()

            if batch_idx % args['log_interval'] == 0:
                print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                    epoch, batch_idx * len(data), len(train_loader.dataset),
                    100. * batch_idx / len(train_loader), loss.item()))


def test(args, model, device, dev_samples):
    model.eval()
    true_y_list = []
    pred_y_list = []
    with torch.no_grad():
        for i in range(len(dev_samples)):
            X, Y = dev_samples[i]

            test_items = LibriItems(X, Y, context=args['context'])
            test_loader = torch.utils.data.DataLoader(test_items, batch_size=args['batch_size'], shuffle=False)

            for data, true_y in test_loader:
                data = data.float().to(device)
                true_y = true_y.long().to(device)  

                #import ipdb; ipdb.set_trace()          
                
                output = model(data)
                pred_y = torch.argmax(output, axis=1)

                pred_y_list.extend(pred_y.tolist())
                true_y_list.extend(true_y.tolist())

    train_accuracy =  accuracy_score(true_y_list, pred_y_list)
    return train_accuracy

def predict(args, model, device, pred_samples):
    model.eval()
    pred_y_list = []
    with torch.no_grad():
      for i in range(len(pred_samples)):
          X = pred_samples[i]
          prediction_items = LibriItems(X, context=args['context'], predict=True) 
          prediction_loader = torch.utils.data.DataLoader(prediction_items, batch_size=args['batch_size'], shuffle=False)
          for data in prediction_loader:  # this for loop should go over test order.csv
              data = data.float().to(device)  
              # import ipdb; ipdb.set_trace()             
              output = model(data)
              pred_y = torch.argmax(output, axis=1)
              pred_y_list.extend(pred_y.tolist())
      predictions = pd.DataFrame()
      predictions['id'] = np.array(range(len(pred_y_list))) 
      predictions['label'] = pred_y_list
      #print(predictions.head)
      # predictions.to_csv("submission.csv", index=False)
    return predictions


def main(args):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu") 
    
    # if this is the first time running a new architecture:
    model = Network(args['context']).to(device)
    # else: load model
    #model= torch.load(saved_model_path+'.pth') 

    print(model)

    optimizer = optim.Adam(model.parameters(), lr=args['lr'])
    criterion = torch.nn.CrossEntropyLoss()
    # If you want to use full Dataset, please pass None to csvpath
    train_samples = LibriSamples(data_path = args['LIBRI_PATH'], shuffle=True, partition="train-clean-100", csvpath=None)#"/content/train_filenames_subset_8192_v2.csv")
    dev_samples = LibriSamples(data_path = args['LIBRI_PATH'], shuffle=True, partition="dev-clean")
    # import ipdb; ipdb.set_trace()

    for epoch in range(1, args['epoch'] + 1):
        train(args, model, device, train_samples, optimizer, criterion, epoch)
        test_acc = test(args, model, device, dev_samples)
        print('Dev accuracy ', test_acc)
        predict_samples = LibriSamples(data_path = args['LIBRI_PATH'], shuffle=False, partition="test-clean", csvpath='test_order.csv', predict=True )
        predictions = predict(args, model, device, predict_samples)
        predictions.to_csv(saved_submission_directory_path+"submission_model3_epoch_"+str(epoch)+".csv", index=False)
        print('prediction stored in submission_model3_epoch_'+str(epoch)+'.csv')
        torch.save(model, saved_model_path+'_epoch_'+str(epoch)+'.pth')
        print('model saved in model'+str(epoch))


    

if __name__ == '__main__':
    args = {
        'batch_size': 2048,
        'context': 30,
        'log_interval': 300,
        'LIBRI_PATH': '/content/hw1p2_student_data',
        'lr': 0.001,
        'epoch': 20
    }
    main(args)
```
